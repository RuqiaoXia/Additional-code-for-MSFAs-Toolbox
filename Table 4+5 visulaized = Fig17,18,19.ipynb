```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

```

# Samson

```python

fig = plt.figure(figsize = (7,5))
ax = fig.add_subplot(111)

x = [0.05,0.1,0.2,0.3]

RMSE = [0.0207,
0.0410,
0.0830,
0.1232]

Entropy = [0.0208,
0.0415,
0.0847,
0.1232]

DFT = [0.0227,
0.0436,
0.0851,
0.1258]

CA = [0.0205,
0.0410,
0.0831,
0.1233]


ax.plot(x,RMSE,'-x',label = 'RMSE')
ax.plot (x,DFT,'x-',label = 'DFT_mag')
ax.plot (x,Entropy,'x-',label = 'RMS_RGB')
ax.plot (x,CA,'x-',label = 'Classification accuarcy')
ax.grid()
ax.legend(loc=2)
ax.set_xlabel('Magnitude of noise',fontsize=15)
ax.set_ylabel("unmixing RMSE",fontsize=15)

fig = plt.figure(figsize = (7,5))
ax = fig.add_subplot(111)

x = [0.05,0.1,0.2,0.3]

RMSE = [0.414,
0.410,
0.415,
0.411]

Entropy = [0.416,
0.415,
0.424,
0.411]

DFT = [0.453,
0.436,
0.425,
0.419]

CA = [0.411,
0.410,
0.416,
0.411]


ax.plot(x,RMSE,'-x',label = 'RMSE')
ax.plot (x,DFT,'x-',label = 'DFT_mag')
ax.plot (x,Entropy,'x-',label = 'RMS_RGB')
ax.plot (x,CA,'x-',label = 'Classification accuarcy')
ax.grid()
ax.legend(loc=1)

ax.set_xlabel('Magnitude of noise',fontsize=15)
ax.set_ylabel("Relative unmixing error",fontsize=15)

```
