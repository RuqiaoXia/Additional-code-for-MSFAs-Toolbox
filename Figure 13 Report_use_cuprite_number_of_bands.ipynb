```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

```

```python
reslut = [0.018942124337789166,
 0.01843858899771092,
 0.017336532253638584,
 0.016805358964470135,
 0.016306448875785848,
 0.016300129469683285,
 0.016194255874768512,
 0.0164635945583785]

time_cost  = [27.624195098876953,
             37.08338522911072,
             109.8728539943695,
              97.62005591392517,
              103.24372386932373,
              116.11940121650696,
              156.48298287391663,
              136.57744216918945
             ]

x = [4,6,10,11,12,13,14,16]
```

```python
 
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, reslut, '-xb', label = 'RMSE')
ax.legend(loc=2)
ax.grid()
ax2 = ax.twinx()
ax2.plot(x, time_cost, '-xr', label = 'Time cost')
ax2.legend(loc=1)

ax.set_xlabel("Number of bands",fontsize=15)
ax.set_ylabel(r"RMSE",fontsize=15)
ax2.set_ylabel(r"Time cost (sec)",fontsize=15)
ax2.set_ylim(0, 200)
ax.set_ylim(0.015,0.02)
ax2.legend(loc=1)
plt.title('Cuprite',fontsize=15)
```

```python
fig = plt.figure()
lns1 = ax.plot(x, reslut, '-b', label = 'RMSE')
lns2 = ax2.plot(x, time_cost, '-r', label = 'Time cost')

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)

ax.grid()
ax.set_xlabel("Number of bands")
ax.set_ylabel(r"RMSE")
ax2.set_ylabel(r"Time cost (sec)")
ax2.set_ylim(0, 200)
ax.set_ylim(0.015,0.02)
ax2.legend(loc=0)
```
