```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
fig = plt.figure(figsize = (7,5))
ax = fig.add_subplot(111)

x = [0,0.05,0.1]

RMSE = [0.116,0.115,0.135]
DFT_mag = [0.098,0.146,0.115]
RMS_RGB = [0.120,0.127,0.135]
DFT_RGB = [0.100,0.104,0.110]
Spectral_unmixing = [7.00E-09,0.021,0.043]



ax.plot(x,RMSE,'-x',label = 'RMSE')

ax.plot (x,DFT_mag,'x-',label = 'DFT_mag')
ax.plot (x,RMS_RGB,'x-',label = 'RMS_RGB')
ax.plot (x,DFT_RGB,'x-',label = 'DFT_RGB')
ax.plot (x,Spectral_unmixing,'x-',label = 'Spectral_unmixing')
ax.grid()
ax.legend(loc=5)

ax.set_xlabel("Magnitue of noise",fontsize=15)
ax.set_ylabel(r"Unmixing RMSE",fontsize=15)


plt.title('Samson',fontsize=15)
```
